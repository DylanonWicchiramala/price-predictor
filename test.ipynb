```python
def col_delta_p(df, columns, ref_column):
    df = df.copy(deep=True)
    reference = df[ref_column]
    for col in columns:
        series = df[col]
        df[col+'_delta'] = (series-reference)/reference
    df.fillna(value=0, inplace=True)
    return df


def time_delta_p(df, columns):
    df = df.copy(deep=True)
    for col in columns:
        series = df[col]
        sft_series = series.shift(periods=1)
        sft_series[0] = df[col].iloc[0]
        df[col+'_delta'] = (series-sft_series)/sft_series
    df.fillna(value=0, inplace=True)
    return df


# df = time_delta_p(price_df, ['Close'])
df = col_delta_p(price_df, ['High','Low','Close'], 'Open')
df
```
